```python
"""
Actividad 2
"""

# Definir el rango de x
x = np.linspace(0, 2*np.pi, 1000)

# Definir las seales senoidales/cosenoidales
sen = np.sin(x)
cos = np.cos(x)

# Definir la seal exponencial
exp = np.exp(-x)

# Sumar las seales para obtener la seal final
seal = sen + cos + exp

# Graficar la seal
plt.figure(figsize=(10, 6))
plt.plot(x, seal, label='senoide1 + senoide2 + exponencial')
plt.title('Grfica de la seal $y = \sin(x) + \cos(x) + e^{-x}$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""
Actividad 3.1
"""

# Crear una seal impulso (delta)
n = np.arange(-50, 51)
impulso = signal.unit_impulse(len(n), 'mid')  # Seal impulso en el centro

# Desplazar la seal impulso
desplazamiento = 10  # Cambia este valor para desplazar la seal
indice_central = len(n) // 2  # ndice del centro de la seal
impulso_desplazado = signal.unit_impulse(len(n), indice_central + desplazamiento)

# Graficar las seales
plt.figure(figsize=(10, 6))
plt.stem(n, impulso, linefmt='C0:', markerfmt='C0o', label='Impulso original')
plt.stem(n, impulso_desplazado, linefmt='C1-', markerfmt='C1.', label='Impulso desplazado')
plt.title('Seal impulso y seal impulso desplazada')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""
Actividad 3.2
"""
# Parmetros de la seal
frecuencia = 5  # Frecuencia de la seal senoidal
duracion = 1.0  # Duracin de la seal en segundos
fs = 100  # Frecuencia de sampleo

# Generar la seal senoidal
t = np.arange(0, duracion, 1/fs)  # Vector de tiempo
senal = np.sin(2 * np.pi * frecuencia * t)  # Seal senoidal

# Realizar la operacin de sampleo
senal_sampleada = senal[::fs//10]  # Samplear la seal cada 10 puntos

# Graficar la seal original y la seal sampleada
plt.figure(figsize=(10, 6))
plt.plot(t, senal, label='Seal original')
plt.stem(t[::fs//10], senal_sampleada, linefmt='C1-', markerfmt='C1o', label='Seal sampleada')
plt.title('Seal senoidal y seal senoidal sampleada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""
Actividad 3.3
"""
# Parmetros de la seal
frecuencia = 5  # Frecuencia de la seal triangular
duracion = 1.0  # Duracin de la seal en segundos
fs = 100  # Frecuencia de sampleo

# Generar la seal triangular
t = np.arange(0, duracion, 1/fs)  # Vector de tiempo
senal = signal.sawtooth(2 * np.pi * frecuencia * t, 0.5)  # Seal triangular

# Realizar la operacin de sampleo
senal_sampleada = senal[::fs//10]  # Samplear la seal cada 10 puntos

# Graficar la seal original y la seal sampleada
plt.figure(figsize=(10, 6))
plt.plot(t, senal, label='Seal original')
plt.stem(t[::fs//10], senal_sampleada, linefmt='C1-', markerfmt='C1o', label='Seal sampleada')
plt.title('Seal triangular y seal triangular sampleada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""
Actividad 3.5
"""
# Parmetros de la seal
frecuencia = 5  # Frecuencia de la seal senoidal
duracion = 1.0  # Duracin de la seal en segundos

# Generar la seal senoidal
t = np.arange(0, duracion, 1/fs)  # Vector de tiempo
senal = np.sin(2 * np.pi * frecuencia * t)  # Seal senoidal

# Realizar la operacin de sampleo y reconstruccin para diferentes frecuencias de sampleo
for fs in [frecuencia, 2*frecuencia, 5*frecuencia]:
    # Samplear la seal
    paso = max(1, int(fs//10))  # Asegurarse de que el paso sea al menos 1
    t_sampleado = t[::paso]  # Samplear t
    senal_sampleada = senal[::paso]  # Samplear la seal

    # Reconstruir la seal a partir del sampleo
    senal_reconstruida = signal.resample(senal_sampleada, len(t))

    # Graficar la seal original, la seal sampleada y la seal reconstruida
    plt.figure(figsize=(10, 6))
    plt.plot(t, senal, label='Seal original')
    plt.stem(t_sampleado, senal_sampleada, linefmt='C1-', markerfmt='C1o', label='Seal sampleada')
    plt.plot(t, senal_reconstruida, label='Seal reconstruida')
    plt.title(f'Seal original, seal sampleada y seal reconstruida (fs = {fs} Hz)')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.legend()
    plt.grid(True)
    plt.show()
```
