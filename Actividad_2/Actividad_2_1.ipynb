```python
"""
Actividad 2
"""

# Definir el rango de x
x = np.linspace(0, 2*np.pi, 1000)

# Definir las señales senoidales/cosenoidales
sen = np.sin(x)
cos = np.cos(x)

# Definir la señal exponencial
exp = np.exp(-x)

# Sumar las señales para obtener la señal final
señal = sen + cos + exp

# Graficar la señal
plt.figure(figsize=(10, 6))
plt.plot(x, señal, label='senoide1 + senoide2 + exponencial')
plt.title('Gráfica de la señal $y = \sin(x) + \cos(x) + e^{-x}$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""
Actividad 3.1
"""

# Crear una señal impulso (delta)
n = np.arange(-50, 51)
impulso = signal.unit_impulse(len(n), 'mid')  # Señal impulso en el centro

# Desplazar la señal impulso
desplazamiento = 10  # Cambia este valor para desplazar la señal
indice_central = len(n) // 2  # Índice del centro de la señal
impulso_desplazado = signal.unit_impulse(len(n), indice_central + desplazamiento)

# Graficar las señales
plt.figure(figsize=(10, 6))
plt.stem(n, impulso, linefmt='C0:', markerfmt='C0o', label='Impulso original')
plt.stem(n, impulso_desplazado, linefmt='C1-', markerfmt='C1.', label='Impulso desplazado')
plt.title('Señal impulso y señal impulso desplazada')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""
Actividad 3.2
"""
# Parámetros de la señal
frecuencia = 5  # Frecuencia de la señal senoidal
duracion = 1.0  # Duración de la señal en segundos
fs = 100  # Frecuencia de sampleo

# Generar la señal senoidal
t = np.arange(0, duracion, 1/fs)  # Vector de tiempo
senal = np.sin(2 * np.pi * frecuencia * t)  # Señal senoidal

# Realizar la operación de sampleo
senal_sampleada = senal[::fs//10]  # Samplear la señal cada 10 puntos

# Graficar la señal original y la señal sampleada
plt.figure(figsize=(10, 6))
plt.plot(t, senal, label='Señal original')
plt.stem(t[::fs//10], senal_sampleada, linefmt='C1-', markerfmt='C1o', label='Señal sampleada')
plt.title('Señal senoidal y señal senoidal sampleada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""
Actividad 3.3
"""
# Parámetros de la señal
frecuencia = 5  # Frecuencia de la señal triangular
duracion = 1.0  # Duración de la señal en segundos
fs = 100  # Frecuencia de sampleo

# Generar la señal triangular
t = np.arange(0, duracion, 1/fs)  # Vector de tiempo
senal = signal.sawtooth(2 * np.pi * frecuencia * t, 0.5)  # Señal triangular

# Realizar la operación de sampleo
senal_sampleada = senal[::fs//10]  # Samplear la señal cada 10 puntos

# Graficar la señal original y la señal sampleada
plt.figure(figsize=(10, 6))
plt.plot(t, senal, label='Señal original')
plt.stem(t[::fs//10], senal_sampleada, linefmt='C1-', markerfmt='C1o', label='Señal sampleada')
plt.title('Señal triangular y señal triangular sampleada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""
Actividad 3.5
"""
# Parámetros de la señal
frecuencia = 5  # Frecuencia de la señal senoidal
duracion = 1.0  # Duración de la señal en segundos

# Generar la señal senoidal
t = np.arange(0, duracion, 1/fs)  # Vector de tiempo
senal = np.sin(2 * np.pi * frecuencia * t)  # Señal senoidal

# Realizar la operación de sampleo y reconstrucción para diferentes frecuencias de sampleo
for fs in [frecuencia, 2*frecuencia, 5*frecuencia]:
    # Samplear la señal
    paso = max(1, int(fs//10))  # Asegurarse de que el paso sea al menos 1
    t_sampleado = t[::paso]  # Samplear t
    senal_sampleada = senal[::paso]  # Samplear la señal

    # Reconstruir la señal a partir del sampleo
    senal_reconstruida = signal.resample(senal_sampleada, len(t))

    # Graficar la señal original, la señal sampleada y la señal reconstruida
    plt.figure(figsize=(10, 6))
    plt.plot(t, senal, label='Señal original')
    plt.stem(t_sampleado, senal_sampleada, linefmt='C1-', markerfmt='C1o', label='Señal sampleada')
    plt.plot(t, senal_reconstruida, label='Señal reconstruida')
    plt.title(f'Señal original, señal sampleada y señal reconstruida (fs = {fs} Hz)')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.legend()
    plt.grid(True)
    plt.show()
```
