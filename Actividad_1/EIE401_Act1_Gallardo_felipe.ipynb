```python
"""
Actividad 2

Funciones : 

f(x) = e^-x^2
f(z) = ysin(x)
"""

#Parametros del vector 
inicio = 0
salto = 0.1 #salto entre datos
final = 25 # final del vector

# Generación del vector
x = np.arange(inicio, final, salto)

# definir Y como constante
Y = 8
f_x = np.exp(-x**2)
f_z = Y*np.sin(x)

# Creamos figuras
plt.subplot(1, 2, 1)
plt.plot(f_x)
plt.title("Gráfica f(x)")
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(f_z)
plt.title("Gráfica f(z)")
plt.grid()
```

```python
"""
a continuacion se solicita generar una funcion y luego aplicarle ruido gausiano
"""
# Creamos una funcion senoidal

tim = np.arange(0, 100, 0.1)
fcos = 3*np.sin(tim)
#print(F_3)

# Definir la media y la desviacion estandar

mu, sig = 0 , 0.1

# Ruido gaussiano

ruido = np.random.normal(mu, sig, 1000)

fcosgaus = fcos + ruido


# Graficamos
# Creamos una figura
fig, axs = plt.subplots(3)

# graficar cos
axs[0].plot(tim, fcos)
axs[0].set_title("Coseno")

# graficar ruido
axs[1].plot(tim, ruido)
axs[1].set_title("Ruido Gaussiano")

# graficar cos
axs[2].plot(tim, fcosgaus)
axs[2].set_title("Coseno con ruido")

fig.tight_layout()

plt.show()
```
